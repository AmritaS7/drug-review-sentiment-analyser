```python
import pandas as pd
import os

# Load full clean dataset using full path
df = pd.read_csv('/Users/amrita/Documents/GitHub/drug-review-sentiment-analyser/data/df_clean.csv')

print(f"Original shape: {df.shape}")

# Keep only columns needed by the app
cols_needed = ['drugName', 'condition', 'review', 
               'review_clean', 'rating', 'sentiment', 'usefulCount']

df_app = df[cols_needed]

print(f"Reduced shape: {df_app.shape}")

# Save compressed
df_app.to_csv('/Users/amrita/Documents/GitHub/drug-review-sentiment-analyser/data/df_app.csv', 
              index=False)

size = os.path.getsize('/Users/amrita/Documents/GitHub/drug-review-sentiment-analyser/data/df_app.csv') / (1024*1024)
print(f"New file size: {size:.1f} MB")
```

```python
import pandas as pd
import os

# Load the app dataset
df_app = pd.read_csv('/Users/amrita/Documents/GitHub/drug-review-sentiment-analyser/data/df_app.csv')

# Save as gzip compressed CSV
df_app.to_csv('/Users/amrita/Documents/GitHub/drug-review-sentiment-analyser/data/df_app.csv.gz', 
              index=False, 
              compression='gzip')

size = os.path.getsize('/Users/amrita/Documents/GitHub/drug-review-sentiment-analyser/data/df_app.csv.gz') / (1024*1024)
print(f"Compressed file size: {size:.1f} MB")
```
